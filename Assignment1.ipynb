## Analyses - Data Cleaning

```python
# imports
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load and display the data and its size
# load csv from local folder
df = pd.read_csv("personality_data.csv")
display(df.head())
print(f"\nThe above data has {len(df.index)} rows and {len(df.columns)} coloumns.")
```

---
The desired rows for analysis are: 2 to 75  
The desired coloumns for analysis are:  
1. sex
2. age
3. data consent
4. latitude
5. longitude
6. question coloumns (labeled BFI-2_n, where 1<=n<=60)
---
Firstly make sure that everyone has consented to the data mining.  
Then clean the data until it has the desired row/column data.

```python
# data cleaning 
# checking data usage consent (coloumn Q93)
consent_data = df.loc[3:,["Q93"]]
display(consent_data.head())
```

```python
# does anyone not consent?
consent_data = (consent_data[consent_data != "Yes, I consent"])
print(f"Number of people who do not consent: {len(consent_data.dropna())}")
```

---
Therefore all participants in the data have consented to the data mining.  
Continue with cleaning rows/coloumns.

```python
# data cleaning
# get desired rows
df = df.drop(1)

# get desired coloumns and new names
cols = ["Sex","age","LocationLongitude","LocationLatitude"]
row0 = ["Sex","Age","Longitude","Latitude"]
newname = ["Sex","Age","Longitude","Latitude"]
for i in range(1,61):
    cols.append(f"BFI-2_{i}")
    row0.append(df.loc[0,f"BFI-2_{i}"].split("-")[1])
    newname.append(f"Q{i}")
df = df.reindex(columns=cols)

# rename the data
df.iloc[0] = row0
df.index = range(1,len(df)+1)
df.columns = newname

display(df.iloc[:5,:10])
```

```python
# check what values are in each coloumn 
# to see if any need to be removed
for i in range(4):
    col = df.iloc[1:,i]
    print(f"{df.iloc[0,i]}: (NaN?: {col.isna().any()})(length: {len(set(col))})")
    print(set(col), "\n")

# questions are likely share the same values
# collect them in a set to check them
question_answers = set()
for i in range(4,len(df.iloc[0])):
    col = df.iloc[1:,i]
    question_answers.update(col)
print(f"Questions: (NaN?: {col.isna().any()})(length: {len(question_answers)})")
print(question_answers)
    
```

---
From the above sets we can conclude:
- Age and Gender coloumns are missing data
- Longitude, Latitude, and Age are continuous data
- Sex and Questions are categorical data
- There are 5 possible responses to a question

```python
# data cleaning
# remove the rows with the missing data
# then check how many people prefer not to say their gender
df = df.dropna()
to_remove = df[df.loc[:,"Sex"] == "Prefer not to say"]
display(to_remove.iloc[:,:10])
print(f"\nThere are {len(to_remove)} people who prefer not to say their gender")
```

```python
# data cleaning
# Since 5 is not much remove them
# then fix index to match
for i in to_remove.index:
    df = df.drop(i)
df.index = range(1,len(df)+1)
print(f"Data Cleaning is done:")
display(df.iloc[:,:10])
```

## Analyses - Patterns and Results
---
Now that the data is clean, graph sex, age, and location distributions to see any patterns.  
Compare this with the distribution of answers across questions, to see if any generalized findings are derivable.

```python
# convert table data to floats for graphing
for col in ["Age", "Longitude", "Latitude"]:
    df.loc[2:,col] = df.loc[2:,col].astype(float)
```

```python
# plot the sex distribution 
sns.histplot(x="Sex" ,data=df.drop(1), hue="Sex" ,palette="PuOr")
plt.title("Sex distribution")
plt.show()
```

```python
# plot the age distribtion
sns.histplot(x="Age" ,data=df.drop(1), bins=10, hue="Age" ,palette="GnBu")
plt.title("Age distribution")
plt.show()
```

```python
# plot a scatter plot of the locations
sns.scatterplot(x="Longitude", y="Latitude",data=df.drop(1), color="tab:purple")
plt.title("Location distribution")
plt.show()
```

Visually derived results:
- Most participants are female
- Particpants are most likely to be age 20 and under age 40
- Clusters of people at (approx longitude, latitude):
    - (0, 51.5)
    - (-4, 50.5)
---
Further explore relations. 

```python
# graph the relation between stress resiliency and age
sns.histplot(y="Q4" , x="Age",data=df.drop(1), hue="Age", palette="PuBuGn" )
plt.title("Relation between stress resiliency and age")
plt.ylabel("Is stress resilient")
plt.show()
```

From the above graph we can conclude:
- Participants age 20 fell into all categories of stress resiliency  
- Participants ages 30 to 60 agree to being stress resilient

```python
# graph the relation between laziness and sex
sns.swarmplot(y="Q8" , x="Sex",data=df.drop(1), hue="Sex", palette="RdBu" )
plt.title("Relation between laziness and sex")
plt.ylabel("Is lazy")
plt.show()
```

From the above graph we can conclude:
- Female participants follow a power law (extremes are less common)
- Male participants are skewed towards disagreeing to being lazy

```python
# graph the relation between location and being sad often
sns.scatterplot(x="Longitude", y="Latitude",data=df.drop(1), hue="Q39", palette="Accent")
plt.title("Relation between often feeling sad and location")
plt.show()
```

From the above graph we can conclude:  
- There is a mix of sadness agreeablility near cluster (-4, 50,5)
- Most participants near cluster (0, 51.5) agree to being often sad (3/4 are sad)
